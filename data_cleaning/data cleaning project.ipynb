```python
import pandas as pd
```

```python
df=pd.read_excel("Customer Call List.xlsx")
df
```

```python
df
df=df.drop_duplicates()
df=df.drop("Not_Useful_Column",axis=1)
#df.strip("First_Name")
```

```python
df
```

```python
df["First_Name"]=df["First_Name"].str.strip()
```

```python
#df["Last_Name"].str.lstrip("...").str.lstrip("/").str.rstrip("_")
df["Last_Name"]=df["Last_Name"].str.strip("123./_")
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace(r'[-/|]','',regex=True)
df
```

```python
df["Phone_Number"]=df["Phone_Number"].astype(str)
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:x[0:3] + '-' + x[3:6] + '-'+ x[6:10])
df["Phone_Number"]=df["Phone_Number"].replace({"nan--":"","Na--":""})
df
```

```python
df[["address","city","pincode"]]=df["Address"].str.split(",",expand=True)
df=df.drop("Address",axis=1)
df
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace("Yes","Y")
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace("No","N")
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace("Yes","Y")
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace("No","N")
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace("N/a","")
```

```python
df=df.fillna("")
df
```

```python
for i in df.index:
    if df.loc[i,"Do_Not_Contact"]=="Y":
        df.drop(i,inplace=True)
```

```python
df
```

```python
for i in df.index:
    if df.loc[i,"Phone_Number"]=="":
        df.drop(i,inplace=True)
```

```python
df=df.reset_index(drop=True)
df
```
