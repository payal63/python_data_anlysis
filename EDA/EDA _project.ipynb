```python


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
customer_churn=pd.read_csv("Customer Churn.csv")
customer_churn
```

```python
customer_churn.info()
```

# for change totalcharges datatype,total_charges have blank value which cant convert into float,so change blank to o then change datatype

```python
customer_churn["TotalCharges"]=customer_churn["TotalCharges"].replace(" ",0)
customer_churn["TotalCharges"]=customer_churn["TotalCharges"].astype(float)
customer_churn.info()
```

# converted 0 and 1 value of SeniorCitizen into "yes" and "no" to easy understand 

```python
customer_churn["SeniorCitizen"]=customer_churn["SeniorCitizen"].replace({0:"No",1:"Yes"})
```

# 26% customer churn out

```python
plt.figure(figsize=(3,4))
ax=sns.countplot(x="Churn",data=customer_churn)
ax.bar_label(ax.containers[0])
plt.title("Count of Customer vs Churn")
```

```python
plt.figure(figsize=(3,4))
gb=customer_churn.groupby(["Churn"])["Churn"].count()
plt.pie(gb,labels=gb.index,autopct="%1.2f%%")
plt.title("Percentage of Customer Churn")
```

# Gender doesn't affect the customer churn ratio

```python
plt.figure(figsize=(3,4))
sns.countplot(x="gender",data=customer_churn,hue="Churn")
plt.title("Gender vs Churn")
```

# used chatgpt to create stack bar chart

```python
grouped = customer_churn.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
percentages = grouped.div(grouped.sum(axis=1), axis=0) * 100  # Calculate percentages

# Step 2: Plot the stacked bar chart with swapped colors
fig, ax = plt.subplots(figsize=(6, 4))  # Adjust figure size
bar1 = ax.bar(percentages.index, percentages['No'], color='blue', label='Churn: No')  # Color changed to blue
bar2 = ax.bar(percentages.index, percentages['Yes'], bottom=percentages['No'], color='orange', label='Churn: Yes')  # Color changed to orange

# Step 3: Add percentage labels
for bar in bar1:  # Add labels for "Churn: No"
    height = bar.get_height()
    if height > 0:
        ax.text(bar.get_x() + bar.get_width() / 2, height / 2, f'{height:.1f}%', ha='center', va='center', color='white')

for bar in bar2:  # Add labels for "Churn: Yes"
    height = bar.get_height() + bar.get_y()
    if bar.get_height() > 0:
        ax.text(bar.get_x() + bar.get_width() / 2, height - bar.get_height() / 2, f'{bar.get_height():.1f}%', ha='center', va='center', color='white')

# Step 4: Customize the chart
plt.title('Senior Citizen vs Churn')
plt.ylabel('Percentage')
plt.xlabel('Senior Citizen',loc="center")
plt.legend(title='Churn')
plt.tight_layout()

# Show the plot
plt.show()

```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
           'TechSupport', 'StreamingTV', 'StreamingMovies']

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10, 6))  # 3 rows, 3 columns
axes = axes.flatten()  # Flatten the axes array for easier indexing

# Create a count plot for each column
for i, col in enumerate(columns):
    sns.countplot(x=col, data=customer_churn, ax=axes[i],hue="Churn")
    axes[i].set_title(f'Count Plot for {col}')
    axes[i].set_xlabel('')  # Remove x-axis label for simplicity
    axes[i].set_ylabel('Count')

# Remove any empty subplot axes
if len(columns) < len(axes):
    for j in range(len(columns), len(axes)):
        fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

# Most customers with PhoneService or MultipleLines do not churn, but multiple lines have slightly higher churn rates.
Fiber optic internet users have higher churn compared to DSL or no internet service.
Customers with services like OnlineSecurity, DeviceProtection, or TechSupport tend to churn less compared to those without these services.
This helps identify which services may influence customer retention.
